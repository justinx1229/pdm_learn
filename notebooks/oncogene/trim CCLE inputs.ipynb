# Prepare Shared Trimmed CCLE/DepMap Inputs

This notebook is a companion to `scripts/data/prepare_trimmed_inputs.py`. The script is the canonical reproduction path; this notebook shows the same raw-data preparation step interactively. The output is `data/DepMap_Trimmed`, which is required by both oncogene and PPI workflows.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().resolve()
while PROJECT_ROOT != PROJECT_ROOT.parent and not (PROJECT_ROOT / "src").exists():
    PROJECT_ROOT = PROJECT_ROOT.parent

SRC_DIR = PROJECT_ROOT / "src"
if str(SRC_DIR) not in sys.path:
    sys.path.insert(0, str(SRC_DIR))

DATA_DIR = PROJECT_ROOT / "data"
RAW_DIR = DATA_DIR / "raw"
INTERMEDIATE_DIR = DATA_DIR / "DepMap_data"
OUTPUT_DIR = DATA_DIR / "DepMap_Trimmed"

from pdm_learn.oncogene import trim_oncogene_input_tables
from pdm_learn.ppi import derive_shared_trimmed_inputs
```

```python
raw_paths = {
    "expression": RAW_DIR / "CCLE_gene_expression.csv",
    "copy_number": RAW_DIR / "CCLE_gene_cn.csv",
    "shrna": RAW_DIR / "shRNA_Broad.csv",
    "mutation": RAW_DIR / "CCLE_gene_mutation.csv",
    "crispr": RAW_DIR / "Avana_gene_effect_20Q3.csv",
}

raw_paths
```

```python
# Run after placing the raw files above, or edit raw_paths to point at your raw exports.
intermediate_paths = trim_oncogene_input_tables(
    raw_paths,
    INTERMEDIATE_DIR,
    matrix_orientation="genes-as-rows",
    align_cell_lines=True,
)

shared_paths = derive_shared_trimmed_inputs(INTERMEDIATE_DIR, OUTPUT_DIR)
shared_paths
```
